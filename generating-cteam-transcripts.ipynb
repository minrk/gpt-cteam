# Generating the "C" Team

This notebook uses [GPT-2](https://github.com/openai/gpt-2) to generate new fake transcripts of the "C" team based on the transcripts of existing episodes.

For when you have run out of real episodes, and need to generate an infinite number of new ones!

Model trained [on colaboratory](https://colab.research.google.com/drive/1Up9uwclJ4wYm9MFJhoI0_BNBqPeuY-vb), derived from [this example](https://colab.research.google.com/drive/1VLG8e7YSEwypxU-noRNhsv5dW4NfTGce) by [Max Woolf](http://minimaxir.com/).

Training data was manual transcripts from YouTube videos for [The "C" Team](http://www.acq-inc.com/cteam), which is only the first six episodes, plus Anchors Aweigh part 2. Thanks to folks who transcribed those episodes! This totally doesn't work with YouTube's auto transcripts, which don't have speaker markers or handle D&D vocabulary at all.

Due to the randomness of model generation, re-running cells will result in fun and different results.

Step 1: Load the trained C-team model in the repo

````python
import gpt_2_simple as gpt2
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='cteam')
````

Define a `generate` function with defaults for generating text from the model with a given prompt.

The default is to generate 250 words (`length`) 4 times (`nsamples`), to see different results.

````python
def generate(prefix, temperature=0.7, nsamples=4,
             batch_size=4, length=250, **kwargs):
    while nsamples % batch_size:
        batch_size -= 1
        
    return gpt2.generate(
        sess,
        run_name='cteam',
        temperature=temperature,
        prefix=prefix,
        nsamples=nsamples,
        batch_size=batch_size,
        length=length,
        **kwargs
    )
````

The text looks like:

```
speaker: says something
other speaker: says something else
```

So we can see what it thinks would happen after Jerry says something by prompting it with `"jerry:"`:

````python
generate("jerry:")
````

Increasing the "temperature" makes the results a bit more random
(gives it more freedom when picking what comes next based on what came before)

````python
generate("jerry:", temperature=1.0, nsamples=1)
````

Reducing temperature results in a lot of repetition

````python
generate("jerry:", temperature=0.1, nsamples=1)
````

What does it think each of the players would say?

````python
generate("ryan:", length=100, nsamples=2)
generate("amy:", length=100, nsamples=2)
generate("kate:", length=100, nsamples=2)
generate("kris:", length=100, nsamples=2)
````

We can also prompt it with a favorite line:

````python
generate("kris: hello, friend", nsamples=3)
````

````python
generate("amy: I do a bite")
````

````python
generate("ryan: is this impressive in d&d?")
````
